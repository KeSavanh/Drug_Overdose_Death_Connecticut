```python
import pandas as pd
```

```python
dict={'drug1':['N','Y','N','N','N','N'],
     'drug2':['Y','N','N','N','N','N'],
     'drug3':['N','N','N','Y','N','N'],
     'drug4':['N','N','N','N','Y','N']}
df=pd.DataFrame(dict)
df
```

```python
df['drugname']=''
```

```python
list=['drug1','drug2','drug3','drug4']
for index, row in df.iterrows():
    for col in list:
        if row[col]=='Y':
            row['drugname']= col
    
```
